# **5. Data Preprocessing**

```python
# Hapus kolom 'Unnamed: 0' dan 'id' dari kedua dataset
df_train = df_train.drop(columns=['Unnamed: 0', 'id'])
df_test = df_test.drop(columns=['Unnamed: 0', 'id'])
print("\nKolom 'Unnamed: 0' dan 'id' telah dihapus.")
```

```python
# Pisahkan kolom kategorikal dan numerik
categorical_cols = df_train.select_dtypes(include=['object']).columns
numerical_cols = df_train.select_dtypes(include=np.number).columns.tolist()

# Hapus target dari list numerik jika ada
if 'satisfaction' in numerical_cols:
    numerical_cols.remove('satisfaction') # Asumsi 'satisfaction' sudah di-encode

# Lakukan encoding
for col in categorical_cols:
    le = LabelEncoder()
    # Fit HANYA pada data train
    df_train[col] = le.fit_transform(df_train[col])
    # Transform data test menggunakan encoder yang sama
    df_test[col] = le.transform(df_test[col])

print("\nEncoding kategorikal selesai.")
display(df_train.head())
```

```python
# Inisialisasi StandardScaler
scaler = StandardScaler()

# Fit HANYA pada data train
df_train[numerical_cols] = scaler.fit_transform(df_train[numerical_cols])

# Transform data test menggunakan scaler yang sama
df_test[numerical_cols] = scaler.transform(df_test[numerical_cols])

print("\nStandarisasi fitur numerik selesai.")
display(df_train.head())
```

```python
# Tentukan direktori output
output_dir = r'C:\Users\Pongo\Downloads\Submission-SML\Submissioin_SML\Eksperimen_SML_Ashim\preprocessing\data_preprocessing'

# Buat direktori jika belum ada
os.makedirs(output_dir, exist_ok=True)

# Simpan train set yang sudah diproses
df_train.to_csv(f'{output_dir}\\preprocessed_train.csv', index=False)

# Simpan test set yang sudah diproses
df_test.to_csv(f'{output_dir}\\preprocessed_test.csv', index=False)

print(f"\nPreprocessed train and test data saved to {output_dir}")
```
